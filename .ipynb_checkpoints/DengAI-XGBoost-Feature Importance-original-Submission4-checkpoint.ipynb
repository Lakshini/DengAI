```python
def preprocess_data(data_path, labels_path=None):
    # load data and set index to city, year, weekofyear
    df = pd.read_csv(data_path, index_col=[0, 1, 2])
    
    # select features we want
    features = ['reanalysis_specific_humidity_g_per_kg', 
                 'reanalysis_dew_point_temp_k', 
                 'station_avg_temp_c', 
                 'station_min_temp_c']
    df = df[features]
    
    # fill missing values
    df.fillna(df.mean(), inplace=True)

    # add labels to dataframe
    if labels_path:
        labels = pd.read_csv(labels_path, index_col=[0, 1, 2])
    
    # separate san juan and iquitos
    sj_train = df.loc['sj']
    iq_train = df.loc['iq']
    
    if(labels_path!=None):
        sj_labels = labels.loc['sj']
        iq_labels = labels.loc['iq']
    
        return sj_train, iq_train,sj_labels,iq_labels
    else:
        return sj_train, iq_train
```

```python
sj_train, iq_train,sj_labels,iq_labels= preprocess_data('data-processed/dengue_features_train.csv',labels_path="data-processed/dengue_labels_train.csv")
```

```python
sj_labels.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(sj_train, sj_labels, test_size=0.2, random_state=123)
```

```python
X_train_iq, X_test_iq, y_train_iq, y_test_iq = train_test_split(iq_train, iq_labels, test_size=0.2, random_state=123)
```

```python
X_train.shape

```

```python
from xgboost import XGBRegressor
regressor_sj = XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, alpha = 10, n_estimators = 10)
sj_best_model = regressor_sj.fit(X_train,y_train)

regressor_iq = XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, alpha = 10, n_estimators = 10)

iq_best_model = regressor_iq.fit(X_train_iq,y_train_iq)

y_pred_sj = regressor_sj.predict(X_test)

y_pred_iq = regressor_iq.predict(X_test_iq)


```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred_sj ))
print("RMSE_sj: %f" % (rmse))
```

```python
rmse = np.sqrt(mean_squared_error(y_test_iq, y_pred_iq))
print("RMSE_iq: %f" % (rmse))
```

```python
from xgboost import XGBRegressor
regressor_sj = XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 1.3, max_depth = 5, alpha = 10, n_estimators = 10)
sj_best_model_all = regressor_sj.fit(sj_train,sj_labels)

regressor_iq = XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 1.5, max_depth = 5, alpha = 10, n_estimators = 10)

iq_best_model_all = regressor_iq.fit(iq_train,iq_labels)

y_pred_sj = regressor_sj.predict(X_test)

y_pred_iq = regressor_iq.predict(X_test_iq)
```

```python
sj_test, iq_test = preprocess_data('data-processed/dengue_features_test.csv')

sj_predictions = sj_best_model_all.predict(sj_test).astype(int)
iq_predictions = iq_best_model_all.predict(iq_test).astype(int)

submission = pd.read_csv("data-processed/submission_format.csv",
                         index_col=[0, 1, 2])

submission.total_cases = np.concatenate([sj_predictions, iq_predictions])
submission.to_csv("data-processed/benchmark.csv")
```
